```python
## named entity

doc8=nlp(u"Apple to build a Hong Kong factory for $6 million")
```

```python
for token in doc8:
  print(token,end=" | ")
```

```python
for entity in doc8.ents:
  print(entity)
  print(entity.label_)
  print(spacy.explain(entity.label_))
  print("\n")
```
