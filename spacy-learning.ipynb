```python
nlp=spacy.load('en_core_web_sm') ## loadng the language library
```

```python
doc=nlp('Tesla is looking at buying U.S. startup for $ 6 million.')
```

```python
for token in doc:
  print(token.text,token.pos_,token.dep_)
```
