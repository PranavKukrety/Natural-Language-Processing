```python
import spacy
```

```python
nlp=spacy.load('en_core_web_sm')
```

```python
from spacy import displacy
```

```python
doc=nlp(u"Apple is going to build a U.K. factory for $6 milllion.")
```

```python
displacy.render(doc,style='dep',options={'distance':110})
```

```python
doc=nlp("Over the last quarter Apple sold nearly 20 thousand iPods for a profit of $6 million.")
```

```python
displacy.render(doc,style='ent')
```
